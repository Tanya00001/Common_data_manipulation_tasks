# Filtering

```python
import pandas as pd
df = pd.read_excel("SalesReport1.xlsx")
df
```

```python
df.columns
```

```python
# to check whether the condition satisfied or not(True OR False)
df["Qtr 1"]>1000
```

```python
# but we don't actually want like this, instead want data of "Qtr 1">1000.
# to do so..
df2 = df.loc[df["Qtr 1"]>1000]
```

```python
df2
```

```python
# data of only "Qtr 1" and "Qtr 2" after filtering..

df3 = df.loc[df["Qtr 1"]> 1000, ["Qtr 1","Qtr 2"]]
```

```python
df3
```

```python
# filtering without using loc..

df[df["Qtr 1"]>1000]
# but can't choose specific columns like df3, because this will be consider as series, which is immutable in nature.
```

```python
# to find out all the data related to the Product "Camembert Pierrot"

df.loc[df["Product"]=="Camembert Pierrot"]
```

```python
# to find only the Customers those have bought the Product "Camembert Pierrot"

df.loc[df["Product"]=="Camembert Pierrot",["Customer"]]
```

### using "isin"

```python
# for multiple conditions within a single column..

df.loc[df["Product"].isin(["Alice Mutton","Camembert Pierrot"])]
```

```python
# using isin for multiple conditions within a single column and to show specific columns after filtering.

df.loc[df["Product"].isin(["Alice Mutton","Camembert Pierrot"]), ["Customer"]]

# using ".shape" can check the number of cols & rows.
```

```python
# to find the data opposite to the specified one..
# using "~" before dataframe name we can find the customers those have bought Product, other than "Camembert Pierrot" and "Alice Mutton".

df.loc[~df["Product"].isin(["Alice Mutton","Camembert Pierrot"]), ["Customer"]]
```

```python
# for applying multiple conditions like "Qtr 1" should be greater than 1000 and the "Product" should include only "Alice Mutton" and "Camembert Pierrot".


df4 = df.loc[(df["Qtr 1"]>1000) & (df["Product"].isin(["Alice Mutton","Camembert Pierrot"]))]
```

```python
df4
```

##### but the problem is that data frame index no. is of the previous one, so we have to reset the index no. ..
## Resetting the index no.

```python
df4.reset_index()
```

```python
df4
```

#### again there are 2 problems..
#### 1. by using reset_index() the previous index no. became a separate column named "index".
#### 2. also, df4 did not change at all.

```python
# to solve these problems.. 

df4.reset_index(inplace = True, drop = True)
```

### here "inplace = True" changed the original data frame(df4) and "drop = True" dropped the unwanted index.

```python
df4
```

##### and this is, what we want.
