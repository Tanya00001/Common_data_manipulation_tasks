# Sorting

```python
import pandas as pd
df = pd.read_excel("SalesReport1.xlsx")
df
```

```python
# sorting data in ascending order by "Customer" name

df.sort_values(by=["Customer"], ascending = True).head(n = 10)
```

```python
# sorting by "Qtr 1" and "Qtr 2" in ascending and descending order respectively.
# firstly, it will start sorting by "Qtr 1", the output of this will be sorted by "Qtr 2".

df.sort_values(by=["Qtr 1","Qtr 2"], ascending = [True,False]).head(n=10)
```

# Conditionally adding columns

```python
# we are adding one column which is "spend", on the basis of condition,
# if "Qtr 2" value is more than 1000 then add it as "More spend", 
# else "Less spend"

import numpy as np
df["spend"] = np.where(df["Qtr 2"]>1000, "More spend", "Less spend")
```

```python
df.tail(n=20)
```

```python
# we are adding one more column which is "spend_X", on the basis of condition,

# if "Qtr 2" value is more than 1500 then add it as "More spend",
# else, check if it is less than 1000 add it as "Less spend",
# else, add it as "Medium spend"

df["spend_X"] = np.where(df["Qtr 4"]>1500, "More spend",(np.where(df["Qtr 4"]<1000,"Less spend","Medium spend")))
```

```python
df.tail(n = 20)
```

```python
# np array example

df["Year"] = np.array(2019)
```
