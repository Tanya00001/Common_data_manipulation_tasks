# Grouping

### in statistical term- S A C
#### S - Split
#### A - Apply
#### C - Combine

```python
# read the data
import pandas as pd
df = pd.read_excel("UberDrives-2016.xlsx")
df
```

### Groupby Summarize

```python
df.groupby("START")["MILES"].agg("mean")

# for every START location, we are calculating the average of MILES.
```

```python
# can also do the same by doing..

df.groupby("START")["MILES"].mean()
```

```python
# shorthand notation

df.groupby("START").mean()

# it worked because in the uber data there is only one column which is of float type("MILES").
# if there will be more than one column which will be of float or int type, the average or mean will be calculated on the basis of all that.
```

```python
# for every START and STOP location, we are calculating the mean and sum of MILES respectively.

df.groupby(["START", "STOP"])["MILES"].agg(["mean","sum"]).head(30)

# Now, they are in list to represent more than one column and more than one respective function.
# A one on one relation b/w column and function.
```

```python
# for each START location finding the mean & the total distance travelled.

df.groupby("START")["MILES"].agg(["mean","sum"]).head(20)
# one to many relation.
```

### for finding most recent and the earliest travelled date and mean distance travelled for each start city.

#### breaking the statement like,
#### 1. have to group it by START city, and have to find the min and max of START_DATE.
#### 2. have to group it by START city, and have to find the mean of MILES. 

```python
# checking the last 5 rows..
df.tail()
```

```python
# the last row is of total, but we do not want this, so..
df = df.iloc[: -1, :]

# means all the rows except the last one and all the columns.
# through that we changed the original data(assigned back to df) as well as we removed the last row.
```

```python
df.tail()
```

```python
# we need to convert the START_DATE in datetime format to work, as it is by default considers as object.
# for that simply do..
#df["START_DATE"] = pd.to_datetime(df["START_DATE"])
#df.dtypes

# if it is already in datetime format then ignore this code.

df["START_DATE"] = pd.to_datetime(df["START_DATE"])
```

```python
# using dictionary we can differentiate that we need "min" & "max" of "START_DATE" and "mean" of "MILES".
# eg: having work on multiple columns with multiple functions to applying on that.

df1 = df.groupby("START").agg({"START_DATE":["min","max"],"MILES":["mean"]})
df1
```

```python
# the columns name are looking very weird, actually they are nested..

df1.columns
```

```python
# also START is working as index
# advantage is that we can do horizontal selecting here, as we usually do vertical selection, here horizontal can be possible.

df1.loc[["Agnew"],:]
# also index no. still work here..
```

```python
# have to reset index for making START as a unique column..
df1.reset_index(inplace = True)
```

```python
df1
# "inplace = True" will reset the index in original data frame(df1)
```

```python
# now, as the columns are nested we will assign them name by our own.

df1.columns = ["START_CITY", "EARLIEST_TRIP", "RECENT_TRIP", "AVERAGE_DISTANCE"]
df1
```
